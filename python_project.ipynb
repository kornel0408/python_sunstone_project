```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group sales by product line and sort
top_products = df.groupby("PRODUCTLINE")["SALES"].sum().sort_values(ascending=False)

# Convert sales values to integers
top_products = top_products.astype(int)

# Define colors: highlight highest (green) and lowest (red), others (blue)
colors = ["blue"] * len(top_products)
colors[0] = "green"  # Highest sales
colors[-1] = "red"   # Lowest sales

# Create bar plot with color mapping
plt.figure(figsize=(10, 5))
bars = plt.bar(top_products.index, top_products.values, color=colors)

# Add values on top of bars
for bar in bars:
    yval = bar.get_height()  # Get bar height
    plt.text(bar.get_x() + bar.get_width()/2, yval, f"{yval:,}", 
             ha='center', va='bottom', fontsize=10, fontweight='bold')

# Add legend manually
plt.legend(handles=[
    plt.Rectangle((0, 0), 1, 1, color="green", label="Highest Sales"),
    plt.Rectangle((0, 0), 1, 1, color="red", label="Lowest Sales")
])

# Formatting
plt.xticks(rotation=20)
plt.title("Top Performing Product Lines")
plt.ylabel("Total Sales")
plt.xlabel("Product Line")

# Convert y-axis labels to integer format
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"{int(x)}"))

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import calendar
import pandas as pd

# Ensure ORDERDATE is in datetime format
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])

# Extract month numbers (handle NaT values safely)
df['Month'] = df['ORDERDATE'].dt.month

# Drop NaN values before applying month names
df = df.dropna(subset=['Month'])  

# Convert month numbers to abbreviated month names
df['MonthName'] = df['Month'].astype(int).apply(lambda x: calendar.month_abbr[x])

# Define a unique color palette for 12 months
colors = sns.color_palette("tab10", 12)

# Create a boxplot with unique colors
plt.figure(figsize=(10, 5))
sns.boxplot(x='MonthName', y='SALES', data=df, hue='MonthName', palette=colors, legend=False)

# Formatting
plt.title("Seasonal Sales Distribution")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.xticks(rotation=0)  # Rotate for better readability
plt.show()
```
